## Parts of Speech (POS) Tagging

**POS Tagging** assigns a part of speech (noun, verb, adjective, etc.) to each word in a sentence based on its definition and context.

- POS tagging is essential for understanding **grammatical structure** of a sentence.
- Useful in downstream tasks like:
    - parsing/syntactic analysis
    - text-to-speech
    - machine translation
    - speech recognition

### Common POS Tags
| Tag | Meaning | Example |
|-----|---------|--------|
| DT | Determiner | The |
| JJ | Adjective | quick, brown, lazy |
| NN | Noun, singular | fox, dog |
| NNS | Noun, plural | cats, bikes |
| VB | Verb, base form | run, jump |
| VBD | Verb, past tense | ran, went|
| VBZ | Verb (3rd person singular present) | jumps |
| IN | Preposition | over |
| RB | Adverb | slowly, everywhere |

### How NLP Systems Perform POS Tagging

| Step | Process             | What Happens                                                              | Example                                            |
| ---- | ------------------- | ------------------------------------------------------------------------- | -------------------------------------------------- |
| 1    | Input Sentence      | NLP system receives raw text                                              | *The cat sits on the mat*                          |
| 2    | Tokenization        | Sentence is split into words (tokens)                                     | [The, cat, sits, on, the, mat]                     |
| 3    | Lexicon Lookup      | Possible POS tags are retrieved from a dictionary                         | cat → noun, sits → verb                            |
| 4    | Context Analysis    | Surrounding words help decide correct tag                                 | book → noun or verb depending on context           |
| 5    | Apply Tagging Model | Algorithm selects most probable tag (Rule-based, HMM, CRF, Neural models) | DET + NOUN + VERB pattern                          |
| 6    | Tag Assignment      | Final POS tag is attached to each token                                   | The/DET cat/NOUN sits/VERB on/IN the/DET mat/NOUN |
| 7    | Output              | Tagged sentence is produced for further NLP tasks                         | Used for parsing, NER, translation                 |




### Statistical Tagging

Uses **probabilities learned from labeled text corpora**.

Example:

```
P(tag | word, previous tag)
```

Algorithms used:

* Hidden Markov Model (HMM)
* Conditional Random Fields (CRF)

The system chooses the tag with **highest probability**.

---
### Machine Learning / Deep Learning Tagging

Modern NLP uses models such as:

* **BiLSTM**
* **Transformer models**
* **BERT**

These models learn **contextual relationships** from large datasets.

Example:

```
The bank of the river
```

Model learns that **bank → noun (river side)**, not financial institution.


````python
import nltk
from nltk import pos_tag, word_tokenize

# Ensure necessary resources are downloaded
nltk.download('punkt_tab')    # tokenization model
nltk.download('averaged_perceptron_tagger_eng')

# Input sentence
sentence = "The quick brown fox jumps over the lazy dog."

# Tokenize and POS tag
tokens = word_tokenize(sentence)
pos_tags = pos_tag(tokens)

# Display result
print("Tokens with POS Tags:")
for token, tag in pos_tags:
    print(f"{token} \t  --- \t {tag}")
````

## Parsing


**Parsing** (also known as syntactic analysis) is the process of analyzing the structure of string of symbols (in natural language or computer code) to understand grammatical structure and word relationships.

- Parsing converts text into a structured format like a parse tree.

- It breaks sentences down based on formal grammar rules to enable computer understanding.

- It analyzes the grammatical structure of a sentence to understand the relationships between words.

### Types of Parsing

| Type | Description |
|------|-------------|
| **Dependency Parsing** | Identifies relationships between "head" words (base-form words which are in a dictionary) and their dependents as a dependency tree|
| **Constituency Parsing** | Breaks down a sentence into sub-phrases forming a hierarchical tree structure |

## Dependency Parsing
**Syntax**:  rules or procedures that determine the structure and format of a language.
- **Syntactic head (nucleus)**: A core, governing word in a phrase or compound that determines its grammatical category and meaning.
    - It is the essential element that can stand alone to represent the entire construction.
    - Examples include the noun in a noun phrase (e.g., **"book"** in *"big red book"*).
- In dependency parsing, each word is mapped to its syntactic head and the relationship is labeled.

**Examples of Dependency Grammar Relations**

| Category | Dependency | Description | Example |
|----------|-----------|------------|---------|
| **Nominal Dependents (Nouns & Modifiers)** | nsubj | Nominal subject of a clause | "**Kevin** can hit the ball." |
|  | obj | Direct object of a verb | "Kevin can hit the **baseball**." |
|  | det | Determiner modifying a noun | "**A** bat" |
|  | amod | Adjectival modifier | "**Large** bird" |
| **Modifiers and Other Relations** | advmod | Adverbial modifier | "She runs **very fast**." |
|  | neg | Negation word | "I do **not** like it." |
|  | prep | Prepositional modifier | "Room **in** my home." |
|  | pobj | Object of preposition | "In my **house**" |
|  | poss | Possession modifier | "**My** house" |
|  | cc | Coordinating conjunction | "Cats **and** dogs" |
|  | conj | Conjunct (linked by conjunction) | "Cats and **dogs**" |
|  | punct | Punctuation | "Hello **,** world **!**" |
|  | aux | Auxiliary verb | "Kevin **can** hit." |




For the sentence: ***"The quick brown fox jumps over the lazy dog"***:
- `fox → nsubj → jumps` --- fox is the nominal (a noun, pronoun, or noun phrase) subject of jumps.
- `The → det → fox` ---The is a determiner of fox.
- `jumps → ROOT` --- jumps is the root (syntactic head or the main action of the sentence) of the sentence.

## Constituency Parsing

**Constituency Parsing** splits a sentence into hierarchical tree structure (**syntactic tree**) representing its grammatical constituents.


| Acronym       | Full Form                                 | Function                                                                      |
| ------------- | ----------------------------------------- | ----------------------------------------------------------------------------- |
| S             | Sentence                                  | The root of the tree; the whole sentence                                      |
| NP            | Noun Phrase                               | Phrase centered around a noun; can act as subject or object                   |
| VP            | Verb Phrase                               | Phrase centered around a verb; can include objects, complements, or modifiers |
| DT            | Determiner                                | Words like "the", "a", "this" that introduce a noun                           |
| JJ            | Adjective                                 | Words that describe a noun, e.g., "quick", "brown", "lazy"                    |
| NN            | Noun, singular                            | Singular noun, e.g., "fox", "dog"                                             |
| VBZ           | Verb, 3rd person singular present         | Verb form for "he/she/it", e.g., "jumps"                                      |
| PP            | Prepositional Phrase                      | Phrase headed by a preposition, e.g., "over the lazy dog"                     |
| (NP under PP) | Noun Phrase inside a Prepositional Phrase | Object of the preposition  |
            





For **"The quick brown fox jumps over the lazy dog"**, the breakdown of the sentence structure:

1. **S → NP + VP + PUNCT**
The sentence `S` consists of a **subject (NP)**, **predicate (VP)**, and punctuation mark (full stop).  
   - **NP (subject)** = "The quick brown fox"  
   - **VP (predicate)** = "jumps over the lazy dog"  
   - **PUNCT** = "." (full stop at the end)

2. **NP → DT + JJ + JJ + NN**  
"The quick brown fox" noun phrase is broken into:  

   - **DT** = "The" (determiner)  
   - **JJ** = "quick" (adjective)  
   - **JJ** = "brown" (adjective)  
   - **NN** = "fox" (head noun)  

3. **VP → VBZ + PP**  
Verb phrase has:  

   - **VBZ** = "jumps" (main verb)  
   - **PP** = "over the lazy dog" (prepositional phrase as part of the predicate)  

4. **PP → IN + NP**  
Prepositional phrase:  
   - **IN** = "over" (preposition)  
   - **NP** = "the lazy dog" (object of the preposition)  

5. **NP → DT + JJ + NN (inside PP)**  
"the lazy dog" is broken into:  

   - **DT** = "the"  
   - **JJ** = "lazy"  
   - **NN** = "dog"  

Now, the syntactic tree is given as:

```
                      S                            
       _______________|__________________________   
      |                         VP               |
      |                _________|___             |  
      |               |             PP           |
      |               |     ________|___         |  
      NP              |    |            NP       |
  ____|__________     |    |     _______|____    |  
 DT   JJ    JJ   NN  VBZ   IN   DT      JJ   NN  .
 |    |     |    |    |    |    |       |    |   |  
The quick brown fox jumps over the     lazy dog  .
```


The tree shows how words group into phrases and how those phrases relate to each other in the sentence.

## Named Entity Recognition (NER)

**NER** identifies and classifies specific entities in text into predefined categories.

### Common NER Categories
| Label | Meaning | Example |
|-------|---------|--------|
| PERSON | People, including fictional | Elon Musk |
| ORG | Organizations, companies | SpaceX |
| GPE | Countries, cities, states | California, USA |
| DATE | Absolute or relative dates | 2002 |

### Why is NER Important?
- **Information Extraction:** Structured info from unstructured text (documents, emails)
- **Applications:** Search engines, chatbots, knowledge graphs
- **Foundation** for question answering, summarization, and more

### How Does NER Work?
- **Rule-Based Methods:** Patterns and dictionaries (capitalization, keywords)
- **Machine Learning:** Supervised models trained on labeled data
- **Deep Learning:** Embeddings (BERT, spaCy) for advanced context understanding

````python
import spacy

# Load spaCy's English NER model
nlp = spacy.load("en_core_web_sm")

# Short input text
text = "Elon Musk founded SpaceX in 2002 in California, USA."

# Process the text
doc = nlp(text)

# Print recognized entities
print("Named Entities, their labels, and explanations:")
for ent in doc.ents:
    print(f"{ent.text} - {ent.label_} ({spacy.explain(ent.label_)})")
````

## Sentiment Analysis

**Sentiment Analysis** determines the emotional tone (positive, negative, neutral) behind text.

### The VADER Lexicon (Base Valence)

- VADER (Valence Aware Dictionary and sEntiment Reasoner) is a lexicon and rule-based sentiment analysis method specifically for social media.
- Valence means ability to affect others. Since it is a lexicon-based tool designed for social media, the valence score is assigned based on human-rated intensity rather than pure logic.
- A valence score is a numerical value assigned to a word that indicates how positive or negative its sentiment is. In a lexicon, VADER assigns valence score $ ∈ [−4,+4]$.
- VADER relies on a dictionary of over 7,500 lexical features (words, emoticons, slang) that were rated by 10 independent human raters on a scale of -4 (Extremely Negative) to +4 (Extremely Positive).
- Neutral words: Removed if they have a mean rating of 0.
- Intensity: Captures both polarity (+/-) and strength (e.g., "okay" = 0.9, "great" = 3.1, "horrible" = -2.5)


Example valence scores:
| Word      | Valence Score | Interpretation         |
| --------- | ------------- | ---------------------- |
| great     | +3.1          | strongly positive      |
| excellent | +3.5          | very strongly positive |
| good      | +1.9          | positive               |
| okay      | +0.9          | slightly positive      |
| bad       | −2.5          | negative               |
| terrible  | −3.5          | very negative          |
| horrible  | −3.7          | extremely negative     |


### VADER Sentiment Scores
| Score | Meaning |
|-------|--------|
| **neg** | Negative sentiment score (0 to 1) |
| **neu** | Neutral sentiment score (0 to 1) |
| **pos** | Positive sentiment score (0 to 1) |
| **compound** | Overall score from -1 (most negative) to +1 (most positive) |

**Real-World Applications:** Twitter analysis, product reviews, financial sentiment, brand monitoring

### VADER Sentiment Calculation
In **VADER**, the **`pos`**, **`neu`**, and **`neg`** scores are **ratios of token sentiment categories** after weighting by the lexicon valence. They are **not simply counts of words**. Let’s compute them for the sentence:

```
"Great job"
```
| Step | Concept | Token / Variable | Formula / Rule | Calculation | Result |
|-----|--------|-----------------|---------------|-------------|--------|
| 1 | Tokenization | Tokens | Split sentence into words | ["great", "job"] | 2 tokens |
| 2 | Lexicon lookup | great | VADER valence score | great → 3.1 | 3.1 |
| 3 | Lexicon lookup | job | VADER valence score | job → 0 | 0 |
| 4 | Positive contribution | great | If valence > 0 → valence + 1 | 3.1 + 1 | **4.1** |
| 5 | Negative contribution | none | If valence < 0 → \|valence - 1\| | No negative word | **0** |
| 6 | Neutral count | job | valence = 0 → neutral token | count = 1 | **1** |
| 7 | Total sentiment weight | total | pos_sum + \|neg_sum\| + neu_count | 4.1 + 0 + 1 | **5.1** |
| 8 | Positive ratio | pos | pos_sum / total | 4.1 / 5.1 | **0.804** |
| 9 | Neutral ratio | neu | neu_count / total | 1 / 5.1 | **0.196** |
| 10 | Negative ratio | neg | \|neg_sum\| / total | 0 / 5.1 | **0.000** |
| 11 | Raw sentiment score | S | Sum of valence scores | 3.1 + 0 | **3.1** |
| 12 | Compound formula | compound | S / √(S² + 15) | 3.1 / √(3.1² + 15) | |
| 13 | Square term | S² | 3.1² | 9.61 | |
| 14 | Add constant | S² + 15 | 9.61 + 15 | 24.61 | |
| 15 | Square root | √(S² + 15) | √24.61 | 4.960 | |
| 16 | Final compound | compound | 3.1 / 4.960 | **0.6249** |

### Final VADER Output

| Metric | Value |
|------|------|
| Positive (pos) | **0.804** |
| Neutral (neu) | **0.196** |
| Negative (neg) | **0.000** |
| Compound | **0.6249** |

````python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk; nltk.download('vader_lexicon')
print(SentimentIntensityAnalyzer().polarity_scores("Great job"))
print(SentimentIntensityAnalyzer().polarity_scores("WoW!, that is amazing."))
print(SentimentIntensityAnalyzer().polarity_scores("WoW!, YOU DID IT AGAIN."))
print(SentimentIntensityAnalyzer().polarity_scores("WoW!, EXCELLENT."))
print(SentimentIntensityAnalyzer().polarity_scores("I am astonished you can manage wearing that outfit."))
````

## Class Task 2 — Sentiment Analysis with Negation

Analyze the sentiment of the following sentences and explain how negation changes the sentiment. Use VADER.

**Sentences:**
1. "I am not unhappy with the results."
2. "This movie is not bad, but it's not great either."
3. "The service was far from excellent."

**Questions to Answer:**
1. What polarity scores are assigned to each sentence?
2. Do the scores align with the actual sentiment expressed? Why or why not?
3. How would you modify the model to better handle such cases?

````python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()

sentences = [
    "I am not unhappy with the results.",
    "This movie is not bad, but it's not great either.",
    "The service was far from excellent."
]

for sentence in sentences:
    scores = sia.polarity_scores(sentence)
    print(f"\nSentence: {sentence}")
    print(f"Scores: {scores}")
````
