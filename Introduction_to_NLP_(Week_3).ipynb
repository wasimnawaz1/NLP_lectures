# 1. Regular Expressions (Regex)

A Regular Expression (Regex) is a powerful tool used in text processing to define search patterns. These patterns can be used for:

   - **Matching**: Identifying specific text patterns in
strings.
   - **Searching**: Finding and extracting specific
sequences of characters.
   - **Replacing**: Modifying parts of text based on defined patterns.
   - **Tokenization**
   - **Named Entity Recognition** (rule-based)
   - **Information Extraction**

Regular expressions are widely used in academia and industry for:
   
1.   Text cleaning, preprocessing, removing unwanted characters (e.g., punctuation), and extracting specific patterns like dates or emails.
2.   Finding names, addresses, or phone numbers in documents.
3.   Creating pipelines for search engines or chatbots.
4.   Validating inputs (e.g., email addresses).
5.   Enable effective search and text analysis.


We can think of regex as a small “domain-specific language” (DSL) for text processing.

- Learning benefits:
   - Faster text searching and validation
   - More concise string manipulation
   - Useful in NLP, data cleaning, unstructured log data parsing to get JSON, CSV etc., web scraping, etc.




### Regular Expressions: Matching Patterns

brackets.

- Special characters (e.g., ., *, +, ?) lose their special meaning inside [].

### Python Regular Expressions
Regex and Python both use backslash \ for special characters. You need to
escape the backslash with another backslash!

   - Example: "\\s+" matches one or more spaces.
   - "\\w+" matches one or more alphanumeric characters.
   - "\\n" in Python is the newline character, not a regex pattern. Use "\\\n" if you want to match a literal backslash followed by n.

#### Use Python's Raw String Notation
   -  Raw strings make it easier to write regex patterns without escaping the
backslash multiple times.
   - Example:
      - r"\d+" matches one or more digits (instead of "\\\d+").
      - r"\bword\b" matches "word" as a whole word by specifying word boundary using \b.

### Capture Groups (Multiple Regex)
Regex pattern

     the (.*)er we (.*), the \1er they \2

(.*): Captures text before "er" and after "we."

\1 and \2: Reuse the captured text in the second segment for validation.

### Non Capture Group
Regex pattern

    (?:some|a few) (people|cats) like some \1



image.png


- Non capture group (?:...):
   - Groups "some" or "a few" without capturing them.
- Capturing Group (people|cats):
   - Captures "people" or "cats" to allow backreference.
- Backreference \1:
   - Reuses the captured group to ensure consistency in the match.


### Regex Categories and Examples



| Category                | Regex        | Meaning                          | Example                          | Explanation |
|------------------------|-------------|----------------------------------|----------------------------------|-------------|
| **Literal**            | `cat`       | Exact match                      | `cat` in "cat dog"               | Matches exact sequence `cat` |
| **Wildcard**           | `.`         | Any character (except newline)   | `c.t` → `cat`, `cut`             | `.` matches any single character |
| **Character Class/Set**    | `[abc]`     | Match a, b, or c                 | `[aeiou]` → `a` in "cat"         | Matches any one listed character |
| **Range**              | `[a-z]`     | Lowercase letters                | `[0-9]` → `5`                    | Matches any character in given range |
| **Negation (Caret ^ first char**           | `[^a-z]`    | NOT lowercase letters            | `[^a-z]` → `A`, `1`, `@`         | Matches anything except lowercase letters |
| **Digit**              | `\d`        | Any digit                        | `\d+` → `123`                    | Equivalent to `[0-9]` |
| **Non-digit**          | `\D`        | Not a digit                      | `\D+` → `abc`                    | Matches anything except digits |
| **Word char**          | `\w`        | Letters, digits, underscore      | `\w+` → `hello_1`                | Equivalent to `[A-Za-z0-9_]` |
| **Non-word**           | `\W`        | Not word character               | `\W` → `@`                       | Matches symbols like `@`, `#`, etc. |
| **Whitespace**         | `\s`        | Space, tab, newline              | `\s+` → spaces                   | Matches whitespace characters |
| **Non-whitespace**     | `\S`        | Not whitespace                   | `\S+` → `word`                   | Matches non-space characters |
| **Start Anchor**       | `^`         | Start of string                  | `^Hello`                         | Matches only if string starts with "Hello" |
| **End Anchor**         | `$`         | End of string                    | `end$`                           | Matches only if string ends with "end" |
| **Word Boundary**      | `\b`        | Whole word boundary              | `\bcat\b`                        | Matches "cat" but not "scatter" |
| **Non-boundary**       | `\B`        | Not a word boundary              | `\Bcat\B`                        | Matches "cat" inside a word (e.g., "scatman") |
| **Zero or More**       | `*`         | 0 or more repetitions            | `go*gle` → `ggle`, `google`      | `o` can appear zero or many times |
| **One or More**        | `+`         | 1 or more repetitions            | `go+gle` → `google`, `gooogle`   | At least one `o` required |
| **Optional**           | `?`         | 0 or 1 occurrence                | `colou?r` → `color`, `colour`    | `u` is optional |
| **Exact Count**        | `{n}`       | Exactly n times                  | `\d{3}` → `123`                  | Matches exactly 3 digits |
| **Range Count**        | `{n,m}`     | Between n and m times            | `\d{2,4}` → `12`, `1234`         | Matches 2 to 4 digits |
| **At Least n**         | `{n,}`      | n or more times                  | `\d{2,}` → `12`, `12345`         | Matches 2 or more digits |
| **Grouping**           | `(abc)`     | Capture group                    | `(ab)+` → `abab`                 | Groups pattern and captures match |
| **Backreference**      | `\1`        | Repeat captured group            | `(cat)\1` → `catcat`             | Repeats previously captured group |
| **Non-capturing Group**| `(?:abc)`   | Group without capturing          | `(?:ab)+`                        | Groups pattern without saving it |
| **OR (pipe)**          | `a\|b`      | Match a OR b                     | `cat\|dog`                       | Matches either "cat" or "dog" |
| **Greedy**             | `.*`        | Match as much as possible        | `<.*>` → `<tag>text</tag>`       | Matches longest possible string |
| **Lazy (non-greedy)**  | `.*?`       | Match as little as possible      | `<.*?>` → `<tag>`                | Matches shortest possible string |
| **Positive Lookahead** | `(?=...)`   | Must be followed by pattern      | `\d(?=kg)` → `5kg`               | Matches digit only if followed by "kg" |
| **Negative Lookahead** | `(?!...)`   | Must NOT be followed by pattern  | `\d(?!kg)`                       | Matches digit not followed by "kg" |
| **Positive Lookbehind**| `(?<=...)`  | Must be preceded by pattern      | `(?<=\$)\d+` → `100` from "$100" | Extracts digits after `$` |
| **Negative Lookbehind**| `(?<!...)`  | Must NOT be preceded by pattern  | `(?<!\$)\d+`                     | Matches digits not preceded by `$` |

```python
### Regex examples covering the key components, using `re.findall` to demonstrate matches and prints both the regex and the results.

import re

examples = [
    # Literal
    (r"cat", "I have a cat and a dog."),
    # Wildcard
    (r"c.t", "cat cut cot cet"),
    # Character Class
    (r"[aeiou]", "cat dog"),
    # Range
    (r"[0-9]", "Room 123"),
    # Negation
    (r"[^a-z]", "Hello 123!"),
    # Digit
    (r"\d+", "My number is 4567."),
    # Non-digit
    (r"\D+", "Room 123"),
    # Word char
    (r"\w+", "hello_1 world!"),
    # Non-word char
    (r"\W+", "hello_1! @world"),
    # Whitespace
    (r"\s+", "Hello world\tGoodbye\nMoon"),
    # Non-whitespace
    (r"\S+", "Hello world\tGoodbye\nMoon"),
    # Start anchor
    (r"^Hello", "Hello world"),
    # End anchor
    (r"world$", "Hello world"),
    # Word boundary
    (r"\bcat\b", "A cat scat scatman"),
    # Non-boundary
    (r"\Bcat\B", "scatman concatenate"),
    # Zero or more
    (r"go*gle", "ggle google gogle gooogle"),
    # One or more
    (r"go+gle", "ggle google gogle gooogle"),
    # Optional
    (r"colou?r", "color colour colouur"),
    # Exact count
    (r"\d{3}", "My pin is 1234"),
    # Range count
    (r"\d{2,4}", "12 123 12345"),
    # At least n
    (r"\d{2,}", "1 12 123 12345"),
    # Grouping
    (r"(ab)+", "abab ab abcd"),
    # Backreference
    (r"(cat)\1", "catcat cat dog"),
    # Non-capturing group
    (r"(?:ab)+", "abab ab abcd"),
    # OR pipe
    (r"cat|dog", "I have a cat and a dog."),
    # Greedy
    (r"<.*>", "<tag>text</tag> <b>bold</b>"),
    # Lazy
    (r"<.*?>", "<tag>text</tag> <b>bold</b>"),
    # Positive lookahead
    (r"\d(?=kg)", "5kg 10lb 7kg"),
    # Negative lookahead
    (r"\d(?!kg)", "5kg 10lb 7kg"),
    # Positive lookbehind
    (r"(?<=\$)\d+", "I paid $100 and $50"),
    # Negative lookbehind
    (r"(?<!\$)\d+", "I paid $100 and 50 dollars")
]

for pattern, text in examples:
    matches = re.findall(pattern, text)
    print(f"Regex: {pattern}\nText: {text}\nMatches: {matches}\n{'-'*50}")
```

# Byte Pair Encoding (BPE) Token Learner

---

## How It Works

1. **Initialize Vocabulary**
   - Start with **all individual characters** as tokens.
   - Example: `{A, B, C, a, b, c}`

---

## Repeat the Process

1. **Find Most Frequent Pair**
   - Look for two **adjacent symbols** (characters or tokens) that occur most frequently in the corpus.
   - Example: If `AB` occurs 10 times and `BC` occurs 8 times → choose `AB`.

2. **Merge the Pair**
   - Combine the pair into a **single token** and add to the vocabulary.
   - Example: `AB` merged → `{A, B, C, AB, a, b, c}`

3. **Replace in Corpus**
   - Replace every occurrence of the merged pair with the **new token** `AB`.

---

## Continue Merging

- Repeat the process until:
  - Desired number of merges `k` is reached **OR**
  - No frequent pairs are left

---

## Example Steps

**Initial Tokens:** `{l, o, w, e, r}`  
**Corpus:** `l o w l o w l o w e r l o w e r`

---

### Step 1: First Merge

- **Most frequent pair:** `lo`  
- **Merge:** `lo`  
- **Updated Vocabulary:** `{l, o, w, e, r, lo}`  
- **Updated Corpus:** `lo w lo w lo w e r lo w e r`

---

### Step 2: Second Merge

- **Most frequent pair:** `low`  
- **Merge:** `low`  
- **Updated Vocabulary:** `{l, o, w, e, r, lo, low}`  
- **Updated Corpus:** `low low low e r low e r`

---

### Step 3: Third Merge

- **Most frequent pair:** `er`  
- **Merge:** `er`  
- **Updated Vocabulary:** `{l, o, w, e, r, lo, low, er}`  
- **Updated Corpus:** `low low lower lower`

---

## Final Tokens

- `{low, er}`  

✅ BPE has successfully merged frequent pairs to form meaningful subword tokens.

# 2. Edit Distance (Levenshtein Distance)

Measures similarity between two strings using:
- Insertions
- Deletions
- Substitutions

## Dynamic Programming Formula

D(i,j) = min(
- D(i-1,j) + 1
- D(i,j-1) + 1
- D(i-1,j-1) + cost
)

---

```python
def levenshtein(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0]*(n+1) for _ in range(m+1)]

    for i in range(m+1):
        dp[i][0] = i
    for j in range(n+1):
        dp[0][j] = j

    for i in range(1, m+1):
        for j in range(1, n+1):
            cost = 0 if s1[i-1] == s2[j-1] else 1
            dp[i][j] = min(
                dp[i-1][j] + 1,
                dp[i][j-1] + 1,
                dp[i-1][j-1] + cost
            )

    return dp[m][n]

print(levenshtein("kitten", "sitting"))
```

### Explanation
- Matrix stores subproblem solutions
- Builds solution bottom-up
- Used in spell correction

# Spell Correction using Edit Distance

Idea:
- Compare input word with dictionary
- Choose minimum edit distance

---

```python
dictionary = ["apple", "banana", "grape", "orange"]
word = "appl"

best = min(dictionary, key=lambda w: levenshtein(word, w))
print("Corrected:", best)
```

# 3. N-gram Language Models

## Types
- Unigram: P(w)
- Bigram: P(w2 | w1)
- Trigram: P(w3 | w1, w2)

## MLE Estimation

Bigram:
P(w2|w1) = Count(w1, w2) / Count(w1)

---

```python
from collections import Counter

corpus = "I love NLP and I love AI".lower().split()

# Unigram
unigrams = Counter(corpus)

# Bigram
bigrams = Counter(zip(corpus[:-1], corpus[1:]))

print("Unigrams:", unigrams)
print("Bigrams:", bigrams)
```

### Explanation
- `Counter` counts frequencies
- `zip()` creates bigrams

# 4. Smoothing Techniques

## Problem: Zero Probability

## Solutions:

| Method | Idea |
|------|------|
| Add-k | Add constant k |
| Good-Turing | Adjust rare counts |
| Kneser-Ney | Advanced smoothing |
| Backoff | fallback to lower n-gram |
| Interpolation | combine probabilities |

---

```python
def add_k_smoothing(count_bigram, count_unigram, V, k=1):
    return (count_bigram + k) / (count_unigram + k * V)

print(add_k_smoothing(2, 5, 10))
```

### Explanation
- Prevents zero probability
- Adds small constant k

# 5. Perplexity & Evaluation

## Perplexity Formula

PP = (1 / P(W))^(1/N)

## Cross Entropy

H = - (1/N) Σ log P(w)

Lower perplexity = better model

---

```python
import math

probs = [0.5, 0.4, 0.6]
N = len(probs)

cross_entropy = -sum(math.log(p) for p in probs) / N
perplexity = math.exp(cross_entropy)

print("Cross Entropy:", cross_entropy)
print("Perplexity:", perplexity)
```

### Explanation
- Log probabilities prevent underflow
- Perplexity measures uncertainty
