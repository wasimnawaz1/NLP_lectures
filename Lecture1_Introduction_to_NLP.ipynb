## Challenges in NLP

### 1. Ambiguity — Words and sentences can have multiple meanings

> *"The chicken is ready to eat."*  
> → The chicken is prepared to eat its food, OR it is cooked and ready for us to eat?

> *"The bank is on the river."*  
> → A financial institution, OR the land along the river?

### 2. Contextuality — Context changes meaning

> *"He opened the door for him."*  
> → Literally opened a door, OR gave him an opportunity/chance

> *"He ran to the bank."*  
> → Financial institution OR the edge of a riverbank

### 3. Sarcasm and Idioms — Machines struggle with non-literal language

> *"Oh, that's just fantastic!"*  
> → Sounds positive, but said sarcastically when something goes wrong

> *"Break a leg!"*  
> → Literally harmful, but actually means wishing someone good luck

## Tokenization

**Tokenization** splits text into smaller units (words or sentences). It is the **first step** in most NLP workflows.

Two types:
- **Word Tokenization** — splits text into individual words
- **Sentence Tokenization** — splits text into individual sentences

Expected output:
```
Word Tokens: ['Cats', 'are', 'running', 'faster', 'than', 'ever', '.', 'They', 'are', 'amazing', 'creatures', '.']
Sentence Tokens: ['Cats are running faster than ever.', 'They are amazing creatures.']
```

````python
#!pip install nltk
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt_tab')
# “Punkt” comes from the German word for period (.). It is a sentence and word tokenizer model that
# intelligently detect sentence boundaries using punctuation like: .   !  ? It is smarter than just splitting on "."

# Input text
text = "Cats are running faster than ever. They are amazing creatures."

# Word Tokenization
word_tokens = word_tokenize(text)

# Sentence Tokenization
sentence_tokens = sent_tokenize(text)

print("Word Tokens:", word_tokens)
print("Sentence Tokens:", sentence_tokens)
````

## Stopword Removal

**Stopwords** are common words like "is", "and", "the", etc., which typically **don't contribute to the meaning** of a sentence. Removing them helps focus on meaningful words.

Expected output:
```
Filtered Words: ['Cats', 'running', 'faster', '.']
```

````python
from nltk.corpus import stopwords
# A large and structured collection of texts used for language research and NLP tasks. (A corpus is a dataset of text.)
# Students learn English by reading books - AI learns language by reading a corpus
nltk.download('stopwords')

# Word tokens
word_tokens = ['Cats', 'are', 'running', 'faster', 'than', 'ever', '.']

# Stopword removal
stop_words = set(stopwords.words('english'))     # loads a list of English stopwords from NLTK and converts the list into a set
# because checking membership in a set is much faster than a list.

# List Comprehension: "Go through each word in word_tokens and keep it only if it is NOT a stopword."
filtered_words = [word for word in word_tokens if word.lower() not in stop_words]

print("Filtered Words:", filtered_words)
````

## Stemming

**Stemming** reduces words to their root forms by **removing suffixes**. It is a heuristic (rule of thumb, shortcut) process, so it may **not always result in actual dictionary words**.

Expected output:
```
Stemmed Words: ['run', 'runner', 'run', 'easili', 'fairli']
```

Notice `'easili'` and `'fairli'` — these are not real words, which is a limitation of stemming.

````python
from nltk.stem import PorterStemmer

# Input words
words = ['running', 'runner', 'runs', 'easily', 'fairly']

# Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in words]

print("Stemmed Words:", stemmed_words)
````

## Porter Stemming Algorithm

**Porter Stemming** was introduced by Martin Porter in **1980**. It applies a sequence of rules iteratively to remove suffixes.

### Key Features
1. **Rule-Based Approach** — applies a sequence of rules to iteratively remove suffixes
2. **Five Phases** — each phase applies specific rules to modify the word
3. **Language-Specific** — designed specifically for the English language

### Steps in Porter Stemming
1. Remove Plurals and -ed/-ing: `caresses → caress`, `running → run`
2. Remove Derivational Suffixes: `national → nation`, `relational → relat`
3. Remove Additional Suffixes: `sensational → sensat`, `logical → logic`
4. Remove -ant, -ence, etc.: `different → differ`, `adoption → adopt`
5. Normalize Final Forms: `caress → caress`, `run → run`

````python
from nltk.stem import PorterStemmer

# Initialize the Porter Stemmer
porter = PorterStemmer()

# List of words to stem
words = ["running", "nationality", "caresses", "flies", "generously", "denationalization"]

# Stem the words
stemmed_words = [porter.stem(word) for word in words]

# Display results
for original, stemmed in zip(words, stemmed_words):
    print(f"Original: {original} --> Stemmed: {stemmed}")
````

## Snowball Stemming Algorithm

**Snowball Stemming** is an improvement over Porter Stemming, also developed by Martin Porter. It is more **robust**, **flexible**, and **supports multiple languages**.

### Steps in Snowball Stemming
1. Normalize words (lowercase, handle special characters)
2. Handle plural forms and common suffixes: `running → run`
3. Language-specific suffix removal: English: `nationality → nation`; French: `nationalité → nation`
4. Iterative refinement through multiple rule passes

### Snowball vs. Porter Comparison

| Feature | Porter Stemming | Snowball Stemming |
|---------|-----------------|-------------------|
| Languages Supported | English only | Multiple languages |
| Flexibility | Less flexible | Highly flexible |
| Performance | Slower | Faster |
| Accuracy | Decent | Better stems, linguistically accurate |

````python
from nltk.stem import SnowballStemmer

# Initialize the Snowball Stemmer for English
snowball = SnowballStemmer(language='english')

# List of words to stem
words = ["running", "nationality", "caresses", "flies", "generously", "denationalization"]

# Stem the words
stemmed_words = [snowball.stem(word) for word in words]

# Display results
for original, stemmed in zip(words, stemmed_words):
    print(f"Original: {original} --> Snowball Stemmed: {stemmed}")
````

## Lemmatization

**Lemmatization** reduces words to their **base form (lemma)** but
ensures the output is a **meaningful word in the dictionary**.

Lemmatization is a an NLP technique that reduces *inflected words* (base words modified by grammatical suffixes (e.g.-s-ed-ing-er)) to their dictionary base form, known as a lemma, e.g.:
- "better" becomes "good,"
- "running" becomes "run"

Unlike stemming, it uses vocabulary, morphological analysis, and context to ensure the resulting base form is valid.

| Word | Stemming Output | Lemmatization Output |
|------|-----------------|---------------------|
| running | run | run |
| easily | easili ❌ | easily ✓ |
| fairly | fairli ❌ | fairly ✓ |

The `pos='v'` argument tells the lemmatizer to treat words as **verbs**.

Expected output:
```
Lemmatized Words: ['run', 'run', 'run', 'easily', 'fairly']
```

````python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

# Input words
words = ['running', 'runner', 'runs', 'easily', 'fairly']

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in words]      # pos means Part Of Speech

print("Lemmatized Words:", lemmatized_words)
````

## Text Representation — Bag of Words (BoW)

**Bag of Words (BoW)** represents text as a **vector of word counts**, disregarding grammar and word order.

- Each row in the resulting matrix represents a document, and each column represents a unique word in the corpus.

- A vocabulary is the set of unique words that appear in a corpus. Cell values represent word counts.

**Limitation:** BoW treats all words equally — it does not account for the relative importance of words.

````python
from sklearn.feature_extraction.text import CountVectorizer

# Sample text data
documents = [
    "Cats are running faster than ever.",
    "Dogs are running too, but not as fast as cats."
]

# Create Bag of Words
vectorizer = CountVectorizer()
bow_matrix = vectorizer.fit_transform(documents)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("BoW Matrix:\n", bow_matrix.toarray())
````

## TF-IDF (Term Frequency–Inverse Document Frequency)

**TF-IDF** assigns importance to words by considering their frequency in a document and **penalizing words that are common across all documents**.

### Term Frequency (TF)
$$TF(t, d) = \frac{\text{Number of times term } t \text{ appears in document } d}{\text{Total number of terms in } d}$$

### Inverse Document Frequency (IDF)
$$IDF(t) = \log\left(\frac{N}{1 + DF(t)}\right)$$

Where **N** = total number of documents, **DF(t)** = number of documents containing the term.

### Final TF-IDF Score
$$TF\text{-}IDF(t, d) = TF(t, d) \times IDF(t)$$

Words appearing in **all** documents get IDF = 0 → TF-IDF = 0 (e.g., "are", "running" appear in both sentences in our example).

````python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Define the sentences
sentences = [
    "Cats are running faster than ever.",
    "Dogs are running too, but not as fast as cats."
]

# Initialize the TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Compute the TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(sentences)

# Convert the TF-IDF matrix into a pandas DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(),
                        columns=vectorizer.get_feature_names_out(),
                        index=["Sentence 1", "Sentence 2"])

# Display the TF-IDF matrix
print(tfidf_df)
````

4. Why do "good" & "sweet" have TF-IDF = 0 across all plays?

Reference: https://web.stanford.edu/~jurafsky/slp3/14.pdf

````python
import math

# Shakespeare corpus
N = 37  # Total number of plays
shakespeare_data = {
    'Romeo':            {'DF': 1},
    'salad':            {'DF': 2},
    'Falstaff':         {'DF': 4},
    'forest':           {'DF': 12},
    'battle':           {'DF': 21},
    'wit':              {'DF': 34},
    'fool':             {'DF': 36},
    'good':             {'DF': 37},
    'sweet':            {'DF': 37}
}

print(f"{'Word':<12} {'DF':>4}   {'IDF':>8}")
print("-" * 30)
for word, vals in shakespeare_data.items():
    idf = math.log(N / (1 + vals['DF']))
    print(f"{word:<12} {vals['DF']:>4}   {idf:>8.4f}")

# TF-IDF for 'Romeo'
tf_romeo  = 16 / 5000
idf_romeo = math.log(N / (1 + 1))
tfidf_romeo = tf_romeo * idf_romeo

print(f"\nTF  for 'Romeo': {tf_romeo:.6f}")
print(f"IDF for 'Romeo': {idf_romeo:.4f}")
print(f"TF-IDF for 'Romeo': {tfidf_romeo:.6f}")
````

## POS Tagging (Parts of Speech)

**POS Tagging** assigns a part of speech (noun, verb, adjective, etc.) to each word in a sentence based on its **definition and context**.

### Why is POS Tagging Important?
- Essential for understanding the **grammatical structure** of a sentence
- Useful in downstream tasks like **parsing**, **text-to-speech**, and **syntactic analysis**

### Common POS Tags
| Tag | Meaning | Example |
|-----|---------|--------|
| DT | Determiner | The |
| JJ | Adjective | quick, brown, lazy |
| NN | Noun | fox, dog |
| VBZ | Verb (3rd person singular present) | jumps |
| IN | Preposition | over |

````python
import nltk
from nltk import pos_tag, word_tokenize

# Ensure necessary resources are downloaded
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Input sentence
sentence = "The quick brown fox jumps over the lazy dog."

# Tokenize and POS tag
tokens = word_tokenize(sentence)
pos_tags = pos_tag(tokens)

# Display result
print("Tokens with POS Tags:")
for token, tag in pos_tags:
    print(f"{token} - {tag}")
````

## Parsing

**Parsing** analyzes the **grammatical structure** of a sentence to understand the relationships between words.

### Types of Parsing

| Type | Description |
|------|-------------|
| **Dependency Parsing** | Identifies relationships between "head" words and their dependents |
| **Constituency Parsing** | Breaks down a sentence into sub-phrases forming a hierarchical tree structure |

## Dependency Parsing

Each word is mapped to its **syntactic head** and the relationship is labeled.

For "The quick brown fox jumps over the lazy dog":
- `fox → nsubj → jumps` (fox is the nominal subject of jumps)
- `The → det → fox` (The is a determiner of fox)
- `jumps → ROOT` (jumps is the root of the sentence)

````python
import spacy

# Load spaCy's English model
nlp = spacy.load("en_core_web_sm")

# Input sentence
sentence = "The quick brown fox jumps over the lazy dog."

# Process the sentence
doc = nlp(sentence)

# Print dependency parsing result
print("Dependency Parsing:")
for token in doc:
    print(f"{token.text:10} {token.dep_:15} {token.head.text:10} {token.head.pos_}")

# Visualizing the dependency parse
from spacy import displacy
displacy.render(doc, style="dep", jupyter=True)
````

## Constituency Parsing

**Constituency Parsing** splits a sentence into **hierarchical tree structures** representing its grammatical constituents.

For "The quick brown fox jumps over the lazy dog":
```
              S
      ________|_______
     NP               VP
   __|___          ___|___
  DT  JJ  JJ VBZ          PP
  |   |   |   |             |
 The quick brown jumps      over
                               |
                              NP
                            ___|___
                           DT      JJ
                           |        |
                          the      lazy
                                    |
                                   NN
                                    |
                                   dog
```

````python
import spacy

# Load spaCy's English model
nlp = spacy.load("en_core_web_sm")

# Add Benepar for constituency parsing
nlp.add_pipe("benepar", config={"model": "benepar_en3"})

# Input sentence
sentence = "The quick brown fox jumps over the lazy dog."

# Process the sentence
doc = nlp(sentence)

# Print the constituency parse tree for the first sentence
for sent in doc.sents:
    print("Constituency Parse Tree:")
    print(sent._.parse_string)

from nltk import Tree
tree = Tree.fromstring(sent._.parse_string)
tree.pretty_print()
````

## Named Entity Recognition (NER)

**NER** identifies and classifies specific entities in text into predefined categories.

### Common NER Categories
| Label | Meaning | Example |
|-------|---------|--------|
| PERSON | People, including fictional | Elon Musk |
| ORG | Organizations, companies | SpaceX |
| GPE | Countries, cities, states | California, USA |
| DATE | Absolute or relative dates | 2002 |

### Why is NER Important?
- **Information Extraction:** Structured info from unstructured text (documents, emails)
- **Applications:** Search engines, chatbots, knowledge graphs
- **Foundation** for question answering, summarization, and more

### How Does NER Work?
- **Rule-Based Methods:** Patterns and dictionaries (capitalization, keywords)
- **Machine Learning:** Supervised models trained on labeled data
- **Deep Learning:** Embeddings (BERT, spaCy) for advanced context understanding

````python
import spacy

# Load spaCy's English NER model
nlp = spacy.load("en_core_web_sm")

# Short input text
text = "Elon Musk founded SpaceX in 2002 in California, USA."

# Process the text
doc = nlp(text)

# Print recognized entities
print("Named Entities, their labels, and explanations:")
for ent in doc.ents:
    print(f"{ent.text} - {ent.label_} ({spacy.explain(ent.label_)})")
````
