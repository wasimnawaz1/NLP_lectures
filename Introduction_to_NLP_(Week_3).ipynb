Regular Expression = A pattern that describes a set of strings
In simple words: Regex is a pattern description language for text.

strings.

### Regular Expressions: Matching Patterns

| [cC]at    | cat, Cat    |

- Ranges using the dash [A-Z]
Dashes specify a range of characters inside the square
brackets.
| Pattern | Matches |
| -------- | -------- |

### Regex Negation

- Special characters (e.g., ., *, +, ?) lose their special meaning inside [].

   - $.$ : Matches any single character (e.g., col.r matches "color" or "colar").

*   Pattern: [!?]\$

### Python Regular Expressions
Regex and Python both use backslash \ for special characters. You need to
escape the backslash with another backslash!

   - Example: "\\s+" matches one or more spaces.
   - "\\w+" matches one or more alphanumeric characters.
   - "\\n" in Python is the newline character, not a regex pattern. Use "\\\n" if you want to match a literal backslash followed by n.

#### Use Python's Raw String Notation
   -  Raw strings make it easier to write regex patterns without escaping the
backslash multiple times.
   - Example:
      - r"\d+" matches one or more digits (instead of "\\\d+").
      - r"\bword\b" matches "word" as a whole word by specifying word boundary using \b.

### ELIZA Chatbot

 - It was created by Joseph Weizenbaum in 1966 and simulated a psychotherapist
 - ELIZA is regex-based, that matches input patterns with regular expressions.
 - It maps identified patterns to empathetic replies.
- It focusses on reflection by encouraging users to elaborate on their statements.


 - Example: Converting statements into questions by identifying user input patterns.
    - Input: "I feel sad."
    - Output: "Why do you feel sad?"  
    - Regex: re.sub(r"I feel (.*)", r"Why do you feel \1?", text).
  
     **((.*) and \1 is an example of capturing groups in regex)**

### Regex Substitutions

1.   Substitutions (Replacing text patterns) in Text Processing
      - Example: Replacing all occurrences of "don't" with "do not".
           - Regex: re.sub(r"\bdon't\b", "do not", text).
2.    Anonymizing data:
       - Example: Masking email addresses in a document.
           -  Regex: re.sub(r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b", "[email removed]", text).

### Capture Groups (Multiple Regex)
Regex pattern

     the (.*)er we (.*), the \1er they \2

(.*): Captures text before "er" and after "we."

\1 and \2: Reuse the captured text in the second segment for validation.



image.png

### Non Capture Group
Regex pattern

    (?:some|a few) (people|cats) like some \1



image.png


- Non capture group (?:...):
   - Groups "some" or "a few" without capturing them.
- Capturing Group (people|cats):
   - Captures "people" or "cats" to allow backreference.
- Backreference \1:
   - Reuses the captured group to ensure consistency in the match.


### Regex Categories and Examples



| Category                | Regex        | Meaning                          | Example                          | Explanation |
|------------------------|-------------|----------------------------------|----------------------------------|-------------|
| **Literal**            | `cat`       | Exact match                      | `cat` in "cat dog"               | Matches exact sequence `cat` |
| **Wildcard**           | `.`         | Any character (except newline)   | `c.t` → `cat`, `cut`             | `.` matches any single character |
| **Character Class/Set**    | `[abc]`     | Match a, b, or c                 | `[aeiou]` → `a` in "cat"         | Matches any one listed character |
| **Range**              | `[a-z]`     | Lowercase letters                | `[0-9]` → `5`                    | Matches any character in given range |
| **Negation (Caret ^ first char**           | `[^a-z]`    | NOT lowercase letters            | `[^a-z]` → `A`, `1`, `@`         | Matches anything except lowercase letters |
| **Digit**              | `\d`        | Any digit                        | `\d+` → `123`                    | Equivalent to `[0-9]` |
| **Non-digit**          | `\D`        | Not a digit                      | `\D+` → `abc`                    | Matches anything except digits |
| **Word char**          | `\w`        | Letters, digits, underscore      | `\w+` → `hello_1`                | Equivalent to `[A-Za-z0-9_]` |
| **Non-word**           | `\W`        | Not word character               | `\W` → `@`                       | Matches symbols like `@`, `#`, etc. |
| **Whitespace**         | `\s`        | Space, tab, newline              | `\s+` → spaces                   | Matches whitespace characters |
| **Non-whitespace**     | `\S`        | Not whitespace                   | `\S+` → `word`                   | Matches non-space characters |
| **Start Anchor**       | `^`         | Start of string                  | `^Hello`                         | Matches only if string starts with "Hello" |
| **End Anchor**         | `$`         | End of string                    | `end$`                           | Matches only if string ends with "end" |
| **Word Boundary**      | `\b`        | Whole word boundary              | `\bcat\b`                        | Matches "cat" but not "scatter" |
| **Non-boundary**       | `\B`        | Not a word boundary              | `\Bcat\B`                        | Matches "cat" inside a word (e.g., "scatman") |
| **Zero or More**       | `*`         | 0 or more repetitions            | `go*gle` → `ggle`, `google`      | `o` can appear zero or many times |
| **One or More**        | `+`         | 1 or more repetitions            | `go+gle` → `google`, `gooogle`   | At least one `o` required |
| **Optional**           | `?`         | 0 or 1 occurrence                | `colou?r` → `color`, `colour`    | `u` is optional |
| **Exact Count**        | `{n}`       | Exactly n times                  | `\d{3}` → `123`                  | Matches exactly 3 digits |
| **Range Count**        | `{n,m}`     | Between n and m times            | `\d{2,4}` → `12`, `1234`         | Matches 2 to 4 digits |
| **At Least n**         | `{n,}`      | n or more times                  | `\d{2,}` → `12`, `12345`         | Matches 2 or more digits |
| **Grouping**           | `(abc)`     | Capture group                    | `(ab)+` → `abab`                 | Groups pattern and captures match |
| **Backreference**      | `\1`        | Repeat captured group            | `(cat)\1` → `catcat`             | Repeats previously captured group |
| **Non-capturing Group**| `(?:abc)`   | Group without capturing          | `(?:ab)+`                        | Groups pattern without saving it |
| **OR (pipe)**          | `a\|b`      | Match a OR b                     | `cat\|dog`                       | Matches either "cat" or "dog" |
| **Greedy**             | `.*`        | Match as much as possible        | `<.*>` → `<tag>text</tag>`       | Matches longest possible string ending with > |
| **Lazy (non-greedy)**  | `.*?`       | Match as little as possible      | `<.*?>` → `<tag>`                | Matches shortest possible string |
| **Positive Lookahead** | `(?=...)`   | Must be followed by pattern      | `\d(?=kg)` → `5kg`               | Matches digit only if followed by "kg" |
| **Negative Lookahead** | `(?!...)`   | Must NOT be followed by pattern  | `\d(?!kg)`                       | Matches digit not followed by "kg" |
| **Positive Lookbehind**| `(?<=...)`  | Must be preceded by pattern      | `(?<=\$)\d+` → `100` from "$100" | Extracts digits after `$` |
| **Negative Lookbehind**| `(?<!...)`  | Must NOT be preceded by pattern  | `(?<!\$)\d+`                     | Matches digits not preceded by `$` |

````python
### Regex examples covering the key components, using `re.findall` to demonstrate matches and prints both the regex and the results.

import re

examples = [
    # Literal
    (r"cat", "I have a cat and a dog."),
    # Wildcard
    (r"c.t", "cat cut cot cet"),
    # Character Class
    (r"[aeiou]", "cat dog"),
    # Range
    (r"[0-9]", "Room 123"),
    # Negation
    (r"[^a-z]", "Hello 123!"),
    # Digit
    (r"\d+", "My number is 4567."),
    # Non-digit
    (r"\D+", "Room 123"),
    # Word char
    (r"\w+", "hello_1 world!"),
    # Non-word char
    (r"\W+", "hello_1! @world"),
    # Whitespace
    (r"\s+", "Hello world\tGoodbye\nMoon"),
    # Non-whitespace
    (r"\S+", "Hello world\tGoodbye\nMoon"),
    # Start anchor
    (r"^Hello", "Hello world"),
    # End anchor
    (r"world$", "Hello world"),
    # Word boundary
    (r"\bcat\b", "A cat scat scatman"),
    # Non-boundary
    (r"\Bcat\B", "scatman concatenate"),
    # Zero or more
    (r"go*gle", "ggle google gogle gooogle"),
    # One or more
    (r"go+gle", "ggle google gogle gooogle"),
    # Optional
    (r"colou?r", "color colour colouur"),
    # Exact count
    (r"\d{3}", "My pin is 1234"),
    # Range count
    (r"\d{2,4}", "12 123 12345"),
    # At least n
    (r"\d{2,}", "1 12 123 12345"),
    # Grouping
    (r"(ab)+", "abab ab abcd"),
    # Backreference
    (r"(cat)\1", "catcat cat dog"),
    # Non-capturing group
    (r"(?:ab)+", "abab ab abcd"),
    # OR pipe
    (r"cat|dog", "I have a cat and a dog."),
    # Greedy
    (r"<.*>", "<tag>text</tag> <b>bold</b>"),
    # Lazy
    (r"<.*?>", "<tag>text</tag> <b>bold</b>"),
    # Positive lookahead
    (r"\d(?=kg)", "5kg 10lb 7kg"),
    # Negative lookahead
    (r"\d(?!kg)", "5kg 10lb 7kg"),
    # Positive lookbehind
    (r"(?<=\$)\d+", "I paid $100 and $50"),
    # Negative lookbehind
    (r"(?<!\$)\d+", "I paid $100 and 50 dollars")
]

for pattern, text in examples:
    matches = re.findall(pattern, text)
    print(f"Regex: {pattern}\nText: {text}\nMatches: {matches}\n{'-'*50}")
````

````python
import re

# Sample text
text = """
Python_3.9 is awesome!
100% Free to try.
Contact us at info@site.com
"""

# Match Digits
digits_pattern = r'\d'
digits_matches = re.findall(digits_pattern, text)
print("Digits Matches:", digits_matches)

# Match Non-Digits
non_digits_pattern = r'\D'
non_digits_matches = re.findall(non_digits_pattern, text)
print("Non-Digits Matches:", ''.join(non_digits_matches))  # Combine to show as a string

# Match Whitespace
whitespace_pattern = r'\s'
whitespace_matches = re.findall(whitespace_pattern, text)
print("Whitespace Matches (count):", len(whitespace_matches))  # Count of whitespaces

# Match Non-Whitespace
non_whitespace_pattern = r'\S'
non_whitespace_matches = re.findall(non_whitespace_pattern, text)
print("Non-Whitespace Matches:", ''.join(non_whitespace_matches))  # Combine to show as a string
````

````python
import re

text = "Contact me at test@email.com or visit https://example.com"

# Extract emails
emails = re.findall(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", text)
print("Emails:", emails)

# Extract URLs
urls = re.findall(r"https?://\S+", text)
print("URLs:", urls)
````

### Code Explanation
- `re.findall()` → finds all matches to capture standard email patterns
- `\S+` → non-space characters
- `?` means “zero or one” occurrence of the preceding character


````python
### Tokenization using Regex

text = "Ali bought 3 apples for $5 on 12/05/2025"

# Tokenization using regex
tokens = re.findall(r"\w+|\$\d+|\d+/\d+/\d+", text)
print(tokens)
````

### Explanation
- `\w+` → words
- `\$\d+` → money, \ is used to escape $ (special character)
- date pattern captures structured dates (using / pattern here).

````python
# NLP Regex Practice Questions

# ==============================
# 1. Setup
# ==============================
import re

print("Regex Examples")


text1 = "My age is 25 and my brother is 30"
print("Digits:", re.findall(r"\d+", text1))

text2 = "NLP is amazing!"
print("Words:", re.findall(r"\b\w+\b", text2))

text3 = "This is a Test sentence"
print("Lowercase words:", re.findall(r"\b[a-z]+\b", text3))


text4 = "Contact: user123@gmail.com, admin@uni.edu"
print("Emails:", re.findall(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", text4))

text5 = "Event on 12/05/2025 and 01/01/2024"
print("Dates:", re.findall(r"\d{2}/\d{2}/\d{4}", text5))

text6 = "Loving #AI and #MachineLearning"
print("Hashtags:", re.findall(r"#\w+", text6))

text7 = "Ali went to Lahore City"
print("Capital words:", re.findall(r"\b[A-Z][a-z]*\b", text7))


text8 = "This is is a repeated repeated word"
print("Repeated words:", re.findall(r"\b(\w+)\s\1\b", text8))

text9 = "I bought 5kg rice and 10kg wheat"
print("Numbers before kg:", re.findall(r"\d+(?=kg)", text9))

text10 = "Items cost $10, $25 and $100"
print("Prices:", re.findall(r"(?<=\$)\d+", text10))

text11 = "sky dry why fly"
print("No vowel words:", re.findall(r"\b[^aeiouAEIOU\s]+\b", text11))


# ==============================
# NLP Tokenization Example
# ==============================
text12 = "Ali bought 3 apples."
print("Tokens:", re.findall(r"\w+|\d+|[^\w\s]", text12))


# ==============================
# URL Extraction
# ==============================
text13 = "Visit https://google.com and http://example.com"
print("URLs:", re.findall(r"https?://\S+", text13))


# ==============================
# Cleaning Text (Replace)
# ==============================
text14 = "This   is    NLP"
cleaned = re.sub(r"\s+", " ", text14)
print("Cleaned:", cleaned)


# ==============================
# Phone Number Validation
# ==============================
text15 = "Call me at 0300-1234567"
print("Phone:", re.findall(r"03\d{2}-\d{7}", text15))


# ==============================
# Words ending with 'ing'
# ==============================
text16 = "I am running and playing while coding"
print("-ing words:", re.findall(r"\b\w+ing\b", text16))


# ==============================
# Exercise Section
# ==============================

print("\n--- Practice Exercises ---")

sample = "Ahmed scored 95 marks on 12/12/2025 and email is ahmed@gmail.com"

# TODO 1: Extract marks
# TODO 2: Extract date
# TODO 3: Extract email

print("Try solving above tasks using regex!")
````
