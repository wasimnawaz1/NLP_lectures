### Text Freature Extraction using N-gram Models

- **Feature**: An individual measurable property or characteristic of an object.


- **Feature Engineering**: The process of using domain knowledge of data to create features that make machine learning algorithms work.


### Freature Extraction

Feature Extraction Is the process of extracting
“features” (or attributes) from *data* in a format supported by machine learning algorithms
- Data can be
   - Text
   - Image
   - Video
   - Audio

- `Scikit-Learn` module for feature extraction
`sklearn.feature_extraction`

**Feature Vector**

- A feature vector contains information of data / object.      
   - It is an $n$-dimensional vector of numerical features representing an object (e.g., frequency counts of words in text, i.e., BoW, TF-IDF, N-grams, etc.).


- For text, features can be *Frequency Counts of Words in a Text* (BoW), TF-IDF, etc.


      Text Features = (a, nice, car)

      Feature Vector = (1, 0, 3)


- **Feature Space**: The $n$-dimensions where your features live.
   - Dimensions of feature vector above is 3.

- For an image, a feature can be a vector of intensity value of pixels.
           
      Image Features = (Pixel 1, Pixel 2, Pixel 3, Pixel 4)
        
      Image Feature Vector = (100, 200, 150, 200)

### Feature Representation of Textual Data


**In Text**
- Features are made up of words (or tokens) / characters

**Transforming String (Text) to Vector**

- Transform documents (or corpus or collection of texts) to features (BoW, TF-IDF, N-grams, etc.)

### N-gram Models

- An **n-gram** is a contiguous sequence of **n items** from a given sample of text
- N-gram can be:
  - **Word-based**
  - **Character-based**
- N represents the **length** of the N-gram

### N-gram Generation from Input Text

**Input Text:** `R u coming?`

Tokenized: `[R, u, coming, ?]`

### Word Uni-grams (N = 1)
Each single word is a feature.

```
Set = {R, u, coming, ?}
```


### Word Bi-grams (N = 2)
Each pair of consecutive words is a feature.

```
Set = {R u, u coming, coming ?}
```

### Word Tri-grams (N = 3)
Each triplet of consecutive words is a feature.

```
Set = {R u coming, u coming ?}
```

### Character Tri-grams (N = 3)
Tokenized characters: `[R, ' ', u, ' ', c, o, m, i, n, g, ?]`

```
Set = {R u, u , co, com, omi, min, ing, ng?}
```

> Note: **space is also a character**

````python
# ─────────────────────────────────────────────
# Word N-gram generation
# ─────────────────────────────────────────────

from nltk import ngrams
import nltk

nltk.download('punkt_tab', quiet=True) # Added to download the required resource

text = "R u coming?"
# Using nltk.word_tokenize for tokenization
tokens_nltk = nltk.word_tokenize(text)

print("Tokens (using nltk.word_tokenize):", tokens_nltk)
print("\nWord Uni-grams (N=1) with nltk.word_tokenize:", list(ngrams(tokens_nltk, 1)))
print("Word Bi-grams  (N=2) with nltk.word_tokenize:", list(ngrams(tokens_nltk, 2)))
print("Word Tri-grams (N=3) with nltk.word_tokenize:", list(ngrams(tokens_nltk, 3)))
````

````python
# ─────────────────────────────────────────────
# Character N-gram generation (manual demo)
# ─────────────────────────────────────────────
text = "R u coming?"
char_tokens = list(text)  # space is also a character

print("Characters:", char_tokens)
print("\nCharacter Tri-grams (N=3):")
char_trigrams = [  ''.join(g) for g in ngrams(char_tokens, 3)   ]
print(char_trigrams)
````

````python
# ─────────────────────────────────────────────
# Example: Text Corpus with Labels (Supervised Learning - Classification)
# ─────────────────────────────────────────────
corpus = [
    "a nice car",
    "a good car",
    "a beautiful car a nice car",
    "a nice black car car"
]

labels = ["Female", "Male", "Female", "Male"]

print("Corpus:")
for i, (text, label) in enumerate(zip(corpus, labels)):
    print(f"  [{i}] '{text}' → {label}")
````

````python
# ─────────────────────────────────────────────
# Example 01 — Binary Weights
# ─────────────────────────────────────────────
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(
    ngram_range=(1, 1),   # unigrams only
    binary=True,          # binary weighting   (1 or 0 no matter how many times a term appear in a doc)
    stop_words=None       # no stop word removal
)

X_binary = vectorizer.fit_transform(corpus)
display(type(X_binary))
df_binary = pd.DataFrame(
    X_binary.toarray(),
    columns=vectorizer.get_feature_names_out()
)
df_binary.index = [f"Doc{i+1} ({l})" for i, l in enumerate(labels)]

print("Document-Term Matrix (Binary Weights):")
df_binary
````

````python
# ─────────────────────────────────────────────
# Example 02 — Term Frequency (Count) Weights
# ─────────────────────────────────────────────
vectorizer_tf = CountVectorizer(
    ngram_range=(1, 1),
    binary=False,         # count / TF weighting
    stop_words=None
)

X_tf = vectorizer_tf.fit_transform(corpus)

df_tf = pd.DataFrame(
    X_tf.toarray(),
    columns=vectorizer_tf.get_feature_names_out()
)
df_tf.index = [f"Doc{i+1} ({l})" for i, l in enumerate(labels)]

print("Document-Term Matrix (Term Frequency Weights):")
df_tf
````

````python
# ─────────────────────────────────────────────
# Example 03 — TF with Stop Words Removal
# ─────────────────────────────────────────────
vectorizer_sw = CountVectorizer(
    ngram_range=(1, 1),
    binary=False,
    stop_words='english'  # removes stop words like 'a'
)

X_sw = vectorizer_sw.fit_transform(corpus)

df_sw = pd.DataFrame(
    X_sw.toarray(),
    columns=vectorizer_sw.get_feature_names_out()
)
df_sw.index = [f"Doc{i+1} ({l})" for i, l in enumerate(labels)]

print("Document-Term Matrix (Stop Words Removed):")
df_sw
````

````python
# ─────────────────────────────────────────────
# Feature Selection — Limit Max Features
# ─────────────────────────────────────────────
vectorizer_sel = CountVectorizer(
    ngram_range=(1, 1),
    binary=False,
    stop_words='english',
    max_features=4        # top-4 most frequent features only
)

X_sel = vectorizer_sel.fit_transform(corpus)

df_sel = pd.DataFrame(
    X_sel.toarray(),
    columns=vectorizer_sel.get_feature_names_out()
)
df_sel.index = [f"Doc{i+1} ({l})" for i, l in enumerate(labels)]

print("Document-Term Matrix (Max Features = 4, Stop Words Removed):")
df_sel
````

````python
# ─────────────────────────────────────────────
# Mixed N-grams (1 to 3), Max Features = 10
# ─────────────────────────────────────────────
vectorizer_mixed = CountVectorizer(
    ngram_range=(1, 3),   # unigrams, bigrams, trigrams
    binary=False,
    stop_words='english',
    max_features=10
)

X_mixed = vectorizer_mixed.fit_transform(corpus)

df_mixed = pd.DataFrame(
    X_mixed.toarray(),
    columns=vectorizer_mixed.get_feature_names_out()
)
df_mixed.index = [f"Doc{i+1} ({l})" for i, l in enumerate(labels)]

print("Document-Term Matrix (Mixed N-grams 1–3, Max Features=10):")
df_mixed
````

````python
# ─────────────────────────────────────────────
# Full Pipeline — Gender Prediction using Ngram featuers and Naive Bayes
# ─────────────────────────────────────────────
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score

# Build pipeline: feature extraction → classifier
pipeline = Pipeline([
    ('vectorizer', CountVectorizer(
        ngram_range=(1, 1),
        stop_words='english',
        max_features=4
    )),
    ('classifier', MultinomialNB())
])

# Fit on training data
pipeline.fit(corpus, labels)

# Predict on a new comment
new_comment = ["a beautiful nice car"]
prediction = pipeline.predict(new_comment)
print(f"New comment: '{new_comment[0]}'")
print(f"Predicted gender: {prediction[0]}")
````
